# Обучение модели

Создадим полезные функции

```python
def make_features(data, max_lag, rolling_size):
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag+1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
        
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_size).mean()
    #return data
```

Проведем обучение модели

### LinearRegression

```python
best_model = None
best_rmse_train = 0
best_rmse_valid = None
best_max_lag = 0
best_rolling_size = 0

for lag in range(1, 301, 10):
    for roll in range(1, 301, 10):
        if roll == 0: continue
        data_main = df.copy()
        make_features(data_main, lag, roll)
        train_valid, test = train_test_split(data_main, test_size=0.1, shuffle=False)
        train, valid = train_test_split(train_valid, test_size=0.2, shuffle=False)
        train = train.dropna()
        
        features_train = train.drop(columns='num_orders')
        target_train = train['num_orders']
        features_valid = valid.drop(columns='num_orders')
        target_valid = valid['num_orders']
        features_test = test.drop(columns='num_orders')
        target_test = test['num_orders']
        
        model = LinearRegression()
        model.fit(features_train, target_train)
        
        predict_train = model.predict(features_train)
        mse_train = mean_squared_error(target_train, predict_train)
        rmse_train = round(sqrt(mse_train), 2)
        
        predict_valid = model.predict(features_valid)
        mse_valid = mean_squared_error(target_valid, predict_valid)
        rmse_valid = round(sqrt(mse_valid), 2)
        if lag == 1:
            best_rmse_valid = rmse_valid
        else:
            if best_rmse_valid > rmse_valid:
                best_rmse_valid = rmse_valid
                best_rmse_train = rmse_train
                best_model = model

                best_max_lag = lag
                best_rolling_size = roll
            
print('Best RMSE = {}; \
      Best RMSE (train) = {}; \
      Best Lag = {}; \
      Best Rolling size = {};'.format(best_rmse_valid, best_rmse_train, best_max_lag, best_rolling_size))

```

### DecisionTreeRegressor

```python
best_model_tree = None
best_rmse_train = 0
best_rmse_valid = None
best_max_lag = 0
best_rolling_size = 0

for depth in range(1, 1000, 50):
    for lag in range(1, 301, 10):
        for roll in range(1, 301, 10):
            if roll == 0: continue
            data_main = df.copy()
            make_features(data_main, lag, roll)
            train_valid, test = train_test_split(data_main, test_size=0.1, shuffle=False)
            train, valid = train_test_split(train_valid, test_size=0.2, shuffle=False)
            train = train.dropna()

            features_train = train.drop(columns='num_orders')
            target_train = train['num_orders']
            features_valid = valid.drop(columns='num_orders')
            target_valid = valid['num_orders']
            features_test = test.drop(columns='num_orders')
            target_test = test['num_orders']

            model = DecisionTreeRegressor(max_depth=depth, random_state=42)
            model.fit(features_train, target_train)

            predict_train = model.predict(features_train)
            mse_train = mean_squared_error(target_train, predict_train)
            rmse_train = round(sqrt(mse_train), 2)

            predict_valid = model.predict(features_valid)
            mse_valid = mean_squared_error(target_valid, predict_valid)
            rmse_valid = round(sqrt(mse_valid), 2)
            if lag == 1:
                best_rmse_valid = rmse_valid
            else:
                if best_rmse_valid > rmse_valid:
                    best_rmse_valid = rmse_valid
                    best_rmse_train = rmse_train
                    best_model_tree = model

                    best_max_lag = lag
                    best_rolling_size = roll
            
print('Best RMSE = {}; \
      Best RMSE (train) = {}; \
      Best Lag = {}; \
      Best Rolling size = {};'.format(best_rmse_valid, best_rmse_train, best_max_lag, best_rolling_size))

```

Лучшая модель - линейная регрессия. Получим результат на тестовой выборке

```python
make_features(df, 201, 221)

train, test = train_test_split(df, test_size=0.1, shuffle=False)
features_test = test.drop(columns='num_orders')
target_test = test['num_orders']

best_model
predict_test = best_model.predict(features_test)
mse_test = mean_squared_error(target_test, predict_test)
rmse_test = round(sqrt(mse_test), 2)
print(rmse_test)
```

Линейная регрессия показала RMSE = 35.32

Лучший показатель `RMSE = 36.49` при гиперпараметрах `lag = 11` и `rolling_size = 181`
